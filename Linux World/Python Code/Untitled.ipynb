```python
import cv2
import numpy as np

# Create a blank image with a white background
width, height = 100, 100
image = np.ones((height, width, 3), dtype=np.uint8) * 255

# Draw a red apple on the image
cv2.circle(image, center=(50, 50), radius=40, color=(0, 0, 255), thickness=-1)

# Get RGB values of each pixel
pixels_rgb = [image[row, col] for row in range(height) for col in range(width)]

print(pixels_rgb)
```
